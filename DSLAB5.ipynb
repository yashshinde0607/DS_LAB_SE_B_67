```python
import pandas as pd
import numpy as np

# Load dataset from UCI repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Column names
cols = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]

df = pd.read_csv(url, names=cols)

print("First 5 rows of dataset:")
print(df.head())

# Filtering rows (PetalLength > 4)
filtered = df[df["PetalLength"] > 4]

print("\nFiltered Data (PetalLength > 4):")
print(filtered)

# Grouping by Species
group = df.groupby("Species")

print("\nMean values by species:")
print(group.mean())

print("\nCount values by species:")
print(group.count())

# NumPy statistics
print("\nStatistics using NumPy")

print("Mean Sepal Length:", np.mean(df["SepalLength"]))
print("Sum Sepal Width:", np.sum(df["SepalWidth"]))
print("Minimum Petal Length:", np.min(df["PetalLength"]))
print("Maximum Petal Width:", np.max(df["PetalWidth"]))
```
