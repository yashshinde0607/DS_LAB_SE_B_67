```python
# program to check the factorial of the given number
num = int(input("Enter a number: "))

fact = 1

for i in range(1, num+1):
    fact = fact * i

print("Factorial is:", fact)
```

```python
# program to check if a number is a palindrome
num = input("Enter a number: ")

if num == num[::-1]:
    print("The number is a Palindrome")
else:
    print("The number is not a Palindrome")
```

```python
# Simple Python program to suggest an engineering field based on interests (without a definition function)

print("Welcome to Engineering Field Suggester")

interest = input("Enter your interest: ")

if interest == "math":
    print("Suggested Field: Mechanical Engineering")

elif interest == "physics":
    print("Suggested Field: Electrical Engineering")

elif interest == "computers":
    print("Suggested Field: Computer Engineering")

elif interest == "biology":
    print("Suggested Field: Biotechnology Engineering")

elif interest == "environment":
    print("Suggested Field: Environmental Engineering")

elif interest == "construction":
    print("Suggested Field: Civil Engineering")

else:
    print("No specific engineering field suggested")
```

```python
# program to check if a number is odd or even
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
# Function to display the multiplication table
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num*i)
```
