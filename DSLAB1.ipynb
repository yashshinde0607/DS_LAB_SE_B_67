```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")
```

```python
# Function to check if a number is a palindrome
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num_to_check_1 = 121
print(is_palindrome(num_to_check_1))

num_to_check_2 = 12321
print(is_palindrome(num_to_check_2))

num_to_check_3 = 12345
print(is_palindrome(num_to_check_3))
```

```python
# Function to check if a number is odd or even
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num_to_check = 10
print(check_odd_even(num_to_check))

num_to_check = 7
print(check_odd_even(num_to_check))
```

```python
# Function to display the multiplication table
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11): # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")

# Example usage:
number_for_table = 7
display_multiplication_table(number_for_table)
```

```python
from experta import *

class StudentFacts(Fact):
    """A fact representing a student's interests."""
    pass

class CareerExpertSystem(KnowledgeEngine):
    """An expert system for suggesting career paths."""

    @Rule(StudentFacts(likes='Maths'), StudentFacts(likes='Physics'))
    def mechanical(self):
        print("Suggested Career Path: Mechanical Engineering")

    @Rule(StudentFacts(likes='Programming'), StudentFacts(likes='Maths'))
    def computer(self):
        print("Suggested Career Path: Computer Engineering")

    @Rule(StudentFacts(likes='Biology'), StudentFacts(likes='Chemistry'))
    def biotech(self):
        print("Suggested Career Path: Biotechnology")

    @Rule(StudentFacts(likes='Circuits'), StudentFacts(likes='Maths'))
    def electronics(self):
        print("Suggested Career Path: Electronics Engineering")

    @Rule(StudentFacts(likes='Art'), StudentFacts(likes='Design'))
    def graphic_design(self):
        print("Suggested Career Path: Graphic Design")

    @Rule(StudentFacts(likes='Writing'), StudentFacts(likes='Reading'))
    def literature(self):
        print("Suggested Career Path: Literature/Journalism")

def main():
    engine = CareerExpertSystem()
    engine.reset()
    print("Welcome to the Career Path Expert System!")
    interests_input = input("Enter your interests separated by commas (e.g., Maths, Physics, Programming): ")
    interests_list = [interest.strip().capitalize() for interest in interests_input.split(',') if interest.strip()]

    for interest in interests_list:
        engine.declare(StudentFacts(likes=interest))

    engine.run()

if __name__ == "__main__":
    main()
```
