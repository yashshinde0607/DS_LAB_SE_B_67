```python
# Take input from the user
print("Enter the elements of the 3x3 matrix row-wise:")

matrix = []
for i in range(3):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

# Assign matrix elements
a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

# Calculate determinant
determinant = (
    a * (e*i - f*h)
    - b * (d*i - f*g)
    + c * (d*h - e*g)
)

# Display result
print("\nMatrix:")
for row in matrix:
    print(row)

print("\nDeterminant of the matrix:")
print(determinant)
```

```python
import numpy as np

# Take input from the user
print("Enter the elements of the 3x3 matrix row-wise:")

elements = []
for i in range(3):
    row = list(map(float, input(f"Enter row {i+1} (3 values separated by space): ").split()))
    elements.append(row)

# Create NumPy array
matrix = np.array(elements)

# Compute determinant
determinant = np.linalg.det(matrix)

# Display results
print("\nMatrix:")
print(matrix)

print("\nDeterminant of the matrix:")
print(determinant)
```

```python
# Finding inverse of the matrix without using function
print("Enter the elements of the 3x3 matrix row-wise:")

m = []
for i in range(3):
    m.append(list(map(float, input(f"Row {i+1}: ").split())))

a, b, c = m[0]
d, e, f = m[1]
g, h, i = m[2]

# Determinant
det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)
print("\nDeterminant:", det)

if det == 0:
    print("Inverse does not exist (determinant is zero).")
else:
    inv = [
        [(e*i - f*h)/det, (c*h - b*i)/det, (b*f - c*e)/det],
        [(f*g - d*i)/det, (a*i - c*g)/det, (c*d - a*f)/det],
        [(d*h - e*g)/det, (b*g - a*h)/det, (a*e - b*d)/det]
    ]

    print("\nInverse of the matrix:")
    for row in inv:
        print(row)

```

```python
# Take input from the user
print("Solve a system of 2 equations of the form: a1*x + b1*y = c1 and a2*x + b2*y = c2")

a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

# Determinant of coefficient matrix
det_A = a1*b2 - a2*b1

if det_A == 0:
    print("The system has no unique solution (determinant is zero).")
else:
    # Using Cramer's Rule
    det_x = c1*b2 - c2*b1
    det_y = a1*c2 - a2*c1

    x = det_x / det_A
    y = det_y / det_A

    print("\nSolution of the system:")
    print(f"x = {x}, y = {y}")
```

```python
import numpy as np

print("Solve a system of 2 equations of the form: a1*x + b1*y = c1 and a2*x + b2*y = c2")

# Take input from the user
a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

# Create coefficient matrix A and constants vector B
A = np.array([[a1, b1],
              [a2, b2]])
B = np.array([c1, c2])

# Check determinant
det_A = np.linalg.det(A)

if det_A == 0:
    print("The system has no unique solution (determinant is zero).")
else:
    # Solve the system
    solution = np.linalg.solve(A, B)
    x, y = solution
    print("\nSolution of the system:")
    print(f"x = {x}, y = {y}")
5
```
