```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# User input
n = int(input("Enter number of values: "))

# Input values
print("Enter the numbers:")
data = []
for i in range(n):
    data.append(int(input()))

data = np.array(data)

# Mean, Median, Mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = stats.mode(data, keepdims=True)[0][0]

# Output
print("\nMean   :", mean_val)
print("Median :", median_val)
print("Mode   :", mode_val)

# Graph
plt.hist(data)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Distribution")
plt.show()

```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# User input
n = int(input("Enter number of values: "))

# Input values
print("Enter the numbers:")
data = []
for i in range(n):
    data.append(int(input()))

data = np.array(data)

# Mean, Median, Mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = stats.mode(data, keepdims=True)[0][0]

# Output
print("\nMean   :", mean_val)
print("Median :", median_val)
print("Mode   :", mode_val)

# ------------------ Graph ------------------

# Histogram + KDE (smooth distribution)
plt.figure()
sns.histplot(data, bins=10, kde=True, color='skyblue')

# Mark mean, median, mode
plt.axvline(mean_val, color='red', linestyle='--', label='Mean')
plt.axvline(median_val, color='green', linestyle='-', label='Median')
plt.axvline(mode_val, color='purple', linestyle=':', label='Mode')

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Distribution (Histogram + Curve)")
plt.legend()
plt.show()

# ------------------ Simple Distribution Curve Only ------------------

plt.figure()
sns.kdeplot(data, fill=True, color='orange')

plt.title("Simple Distribution Curve")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np              # NumPy
from scipy.stats import bernoulli   # SciPy
import matplotlib.pyplot as plt     # Matplotlib

# User input
n = int(input("Enter number of coin tosses: "))

# SciPy → generate random coin tosses
# 0 = Heads, 1 = Tails
tosses = bernoulli.rvs(p=0.5, size=n)

# NumPy → calculations
arr = np.array(tosses)
heads = np.count_nonzero(arr == 0)
tails = np.count_nonzero(arr == 1)

prob_heads = heads / n
prob_tails = tails / n

# Output
print("\nResults:")
print("Total Tosses:", n)
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)

# Matplotlib → visualization
plt.figure()
plt.bar(['Heads', 'Tails'], [heads, tails])
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.title("Coin Toss Simulation")
plt.show()
```
