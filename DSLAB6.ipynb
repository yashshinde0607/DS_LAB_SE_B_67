```python
# Experiment 06: Titanic Data Analysis and Visualization

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display files in directory
print("Files in directory:", os.listdir())

# Load dataset (CSV format)
df = pd.read_csv("Titanic-Dataset(1).xls")

# Display first 5 rows
print("\nFirst 5 Records:")
print(df.head())

# Display dataset information
print("\nDataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# ------------------ Data Preprocessing ------------------

# Fill missing values in Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values in Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin column due to many missing values
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)

# Convert categorical data into numerical form
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# ------------------ Data Visualization ------------------

# Survival count
plt.figure()
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# Survival based on Gender
plt.figure()
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival based on Gender")
plt.xlabel("Gender (0 = Male, 1 = Female)")
plt.ylabel("Count")
plt.show()

# Survival based on Passenger Class
plt.figure()
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival based on Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()

# Age distribution
plt.figure()
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Correlation Heatmap
plt.figure()
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
