```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display files in directory
print("Files in directory:", os.listdir())

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Display first 5 rows
print("\nFirst 5 Records:")
print(df.head())

# Dataset info
print("\nDataset Info:")
print(df.info())

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# ------------------ Data Preprocessing ------------------

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin column
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)

# Convert categorical to numeric
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# ------------------ Visualization ------------------

# 1. Total Survival Count
plt.figure()
sns.countplot(x='Survived', data=df)
plt.title("Total Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# ------------------ Gender-based Survival ------------------

# Separate data
male_df = df[df['Sex'] == 0]
female_df = df[df['Sex'] == 1]

# Male survival chart
plt.figure()
sns.countplot(x='Survived', data=male_df)
plt.title("Survival Count for Male Passengers")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Male Passengers")
plt.show()

# Female survival chart
plt.figure()
sns.countplot(x='Survived', data=female_df)
plt.title("Survival Count for Female Passengers")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Female Passengers")
plt.show()

# ------------------ Passenger Class-based Survival ------------------

# Class 1 (First Class)
class1_df = df[df['Pclass'] == 1]

plt.figure()
sns.countplot(x='Survived', data=class1_df)
plt.title("Survival in First Class (Pclass = 1)")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# Class 2 (Second Class)
class2_df = df[df['Pclass'] == 2]

plt.figure()
sns.countplot(x='Survived', data=class2_df)
plt.title("Survival in Second Class (Pclass = 2)")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# Class 3 (Third Class)
class3_df = df[df['Pclass'] == 3]

plt.figure()
sns.countplot(x='Survived', data=class3_df)
plt.title("Survival in Third Class (Pclass = 3)")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# ------------------ Age Distribution ------------------

plt.figure()
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# ------------------ Correlation Heatmap ------------------

plt.figure()
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
